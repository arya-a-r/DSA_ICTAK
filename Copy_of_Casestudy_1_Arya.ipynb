```python
#A grocery store wants a program to calculate the total bill for a customer.
#a. Store the prices of items in a dictionary (e.g., {'apple': 10, 'milk': 50})
#b. Take the items purchased and their quantities as input.
# c. Calculate and display the total bill amount.




items= {'apple': 100.00, 'milk': 30.00, 'curd': 15.50, 'butter': 20.75,'lays':10}
p_qty = {}
total_bill = 0.00
for i in range(1,21):
    item = input("\nEnter the item you want to purchase (or type 'f' to finish): ").lower()
    if item == "f":
        break
    if item not in items:
        print(f"Item not found in store.\nAvailable items are:{list(items.keys())}")
        continue
    else:
        quantity = int(input(f"\nEnter the quantity of {item} "))
        if quantity >= 0:
           p_qty[item] = quantity
        else:
            print(f"Invalid quantity. Quantity cannot be negative for {item}")


print("\n...........Bill..........")
print("\nItem, Price, and Quantity:\n...........................")
for item, price in items.items():
    if item in p_qty:
      quantity = p_qty.get(item, 0)
      total=price*quantity
      print(f"{item}: Rs.{price:.2f} x {quantity} = Rs.{total:.2f}")
      total_bill += total

print(".................................")
print(f"Total Bill: Rs.{total_bill:.2f}")
b=round(total_bill)
print(f"\n Amount to paid :Rs.{b:.2f}")
print(".................................")
print("Thank you for shopping with us!")


```

```python
#A teacher wants a program to calculate grades for students based on their marks.
#a. Store marks for 5 subjects in a list.
#b. Calculate the average and determine the grade:
# 90+ -> A
# 70-89 -> B
# 50-69 -> C
# <50 -> F
# Display the average and the grade.


def grade_fn (average):
    if average >= 90:
        return 'A'
    elif average >= 70:
        return 'B'
    elif average >= 50:
        return 'C'
    else:
        return 'F'

subjects=['English','Hindi','Math','Social','Malayalam']
marks = []
for subject in subjects:
    mark = float(input(f"\nEnter marks for {subject} out of 100: "))
    while mark < 0 or mark > 100:
        print("Invalid marks. Marks should be between 0 and 100.")
        mark = float(input(f"\nEnter marks for {subject} out of 100: "))
    marks.append(mark)
average = sum(marks) / len(marks)
grade = grade_fn(average)
print(f"\nAverage of Marks: {average:.2f}")
print(f"Grade: {grade}")
```
